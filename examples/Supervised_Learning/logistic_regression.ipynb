# Logistic Regression

This notebook demonstrates binary logistic regression using
`rice_ml.Supervised_Learning.LogisticRegression`, which is trained with the
`GradientDescent` helper.

```python
import sys
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

# Ensure local package imports work when notebook runs from examples/.
repo_root = Path.cwd()
src_path = repo_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from rice_ml.Supervised_Learning import LogisticRegression
```

```python
# Apply a simple linearly-separable binary dataset
rng = np.random.default_rng(7)
X0 = rng.normal(loc=(-1.5, -1.0), scale=0.7, size=(100, 2))
X1 = rng.normal(loc=(1.5, 1.0), scale=0.7, size=(100, 2))
X = np.vstack([X0, X1])
y = np.hstack([np.zeros(X0.shape[0]), np.ones(X1.shape[0])])

print(X.shape, y.shape)
```

```python
# Train logistic regression
model = LogisticRegression(eta=0.15, epochs=500, random_state=7)
model.train(X, y)

y_hat = model.predict(X)
acc = np.mean(y_hat == y)
print(f"Training accuracy: {acc:.3f}")
print(f"Initial loss: {model.loss_history_[0]:.4f}")
print(f"Final loss:   {model.loss_history_[-1]:.4f}")
```

```python
# Plot training loss
plt.figure(figsize=(6, 3.5))
plt.plot(model.loss_history_)
plt.title("Logistic Regression Training Loss")
plt.xlabel("Iteration")
plt.ylabel("Binary cross-entropy")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Visualize data + learned decision boundary (p = 0.5)
plt.figure(figsize=(6, 5))
plt.scatter(X[y == 0, 0], X[y == 0, 1], alpha=0.7, label="Class 0")
plt.scatter(X[y == 1, 0], X[y == 1, 1], alpha=0.7, label="Class 1")

x_min, x_max = X[:, 0].min() - 1.0, X[:, 0].max() + 1.0
y_min, y_max = X[:, 1].min() - 1.0, X[:, 1].max() + 1.0
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 250),
    np.linspace(y_min, y_max, 250),
)
grid = np.c_[xx.ravel(), yy.ravel()]
probs = model.predict_proba(grid).reshape(xx.shape)

# The p=0.5 contour is the classification boundary.
plt.contour(xx, yy, probs, levels=[0.5], linewidths=2)

plt.title("Logistic Regression Decision Boundary")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
